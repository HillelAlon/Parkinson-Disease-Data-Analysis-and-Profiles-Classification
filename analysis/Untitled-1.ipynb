```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import logging
from scipy.stats import poisson

# --- FUNCTION DEFINITIONS ---

def setup_logging():
    """
    Configures a professional logging system.
    Replaces print statements with structured logs.
    """
    logging.basicConfig(
        level=logging.INFO, 
        format='%(asctime)s - %(levelname)s - %(message)s'
    )
    return logging.getLogger(__name__)

def load_parkinson_data(file_path):
    """
    Loads the dataset and sets PatientID as the index.
    """
    try:
        data = pd.read_csv(file_path, index_col='PatientID')
        logger.info(f"Data loaded successfully. Records: {len(data)}, Features: {data.shape[1]}")
        return data
    except FileNotFoundError:
        logger.error(f"File not found at: {file_path}")
        return None

# --- EXECUTION ---
logger = setup_logging()
data_path = '../data/parkinsons_cleaned.csv'
df = load_parkinson_data(data_path)

if df is not None:
    display(df.head())
```
